# Parte 2: Exploración de variables categóricas

Las columnas de texto (sectores, países, etapas) no aparecen en `.describe()`. Para explorarlas usamos otros métodos.

### `.value_counts()` — frecuencia de cada categoría

`.value_counts()` cuenta cuántas veces aparece cada valor en una columna. Es el equivalente a un "¿cuántos hay de cada uno?" muy rápido.

**Pregunta:** ¿qué diferencia hay entre `.value_counts()` y `.nunique()`? ¿Cuándo usarían cada uno?

```python
print("Distribución por sector principal:")
print(df['sector_principal'].value_counts())
print()
print("Distribución por país:")
print(df['pais'].value_counts())
print()
print("Distribución por etapa de financiamiento:")
print(df['etapa'].value_counts())
print()
print("Distribución por modelo de negocio:")
print(df['modelo_negocio'].value_counts())
```

### `.isnull().sum()` — conteo de valores faltantes

Antes de visualizar, conviene tener claro exactamente **cuántos nulos hay por columna**. `.isnull()` devuelve `True/False` por celda, y `.sum()` los suma (en Python, `True` vale 1).

**Pregunta:** ¿qué estrategia usarían para la columna `fondeo_usd_mm` que tiene muchos nulos? ¿Eliminar filas, imputar con la mediana, o ignorar el nulo según el análisis?

```python
nulos = df.isnull().sum()
nulos_pct = (df.isnull().sum() / len(df) * 100).round(1)
resumen_nulos = pd.DataFrame({'nulos': nulos, 'porcentaje %': nulos_pct})
resumen_nulos[resumen_nulos['nulos'] > 0].sort_values('nulos', ascending=False)
```

# Parte 4: Barplots — comparar categorías

Un **barplot** responde preguntas del tipo: *"¿cuánto/cuántos hay de cada categoría?"*

Vamos a usar `sns.countplot()` para contar y `sns.barplot()` para agregar estadísticas.

### `sns.countplot()` — cantidad de startups por sector

`sns.countplot()` cuenta automáticamente las filas de cada categoría y las grafica. No necesitamos calcular nada antes, seaborn lo hace por nosotros.

- `order=` controla el orden de las barras (de mayor a menor es lo más legible)
- `palette=` define la paleta de colores

**Pregunta:** ¿qué conclusión de negocio sacarían de este gráfico? ¿Dice algo sobre el ecosistema de startups LATAM?

```python
orden = df['sector_principal'].value_counts().index

plt.figure(figsize=(10, 5))
sns.countplot(
    data=df,
    x='sector_principal',
    order=orden,
    palette='Set2'
)
plt.title('Cantidad de startups por sector', fontsize=14)
plt.xlabel('Sector')
plt.ylabel('Cantidad de startups')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

### `sns.barplot()` — ARR promedio por etapa de financiamiento

`sns.barplot()` muestra el **promedio** de una variable numérica (y) para cada categoría (x). Las barras de error que aparecen representan el intervalo de confianza.

- `estimator=np.mean` — calcula el promedio (es el default, pero lo escribimos explícito para que quede claro)
- `order=` — ordenamos las etapas de menor a mayor madurez: tiene más sentido narrativo

**Pregunta de interpretación:** ¿qué relación esperaban ver entre etapa y ARR? ¿El gráfico confirma esa hipótesis?

```python
import numpy as np

orden_etapa = ['Pre-seed', 'Seed', 'Serie A', 'Serie B']

plt.figure(figsize=(9, 5))
sns.barplot(
    data=df,
    x='etapa',
    y='arr_millones',
    order=orden_etapa,
    palette='Blues_d',
    estimator=np.mean
)
plt.title('ARR promedio por etapa de financiamiento', fontsize=14)
plt.xlabel('Etapa')
plt.ylabel('ARR promedio (millones USD)')
plt.tight_layout()
plt.show()
```

### `sns.barplot()` — Fondeo promedio por país

Ahora comparamos fondeo promedio entre países. Agregamos `.groupby()` para calcular nosotros mismos el valor antes de graficar, lo que nos da más control.

**Pregunta:** ¿qué diferencia hay entre dejar que `sns.barplot()` calcule el promedio automáticamente vs calcularlo nosotros con `.groupby().mean()`? ¿Cuándo conviene cada enfoque?

```python
fondeo_pais = (
    df.groupby('pais')['fondeo_usd_mm']
    .mean()
    .round(2)
    .sort_values(ascending=False)
    .reset_index()
)
fondeo_pais.columns = ['pais', 'fondeo_promedio_mm']

plt.figure(figsize=(10, 5))
sns.barplot(
    data=fondeo_pais,
    x='pais',
    y='fondeo_promedio_mm',
    palette='Set3'
)
plt.title('Fondeo promedio por país (USD M)', fontsize=14)
plt.xlabel('País')
plt.ylabel('Fondeo promedio (millones USD)')
plt.xticks(rotation=30, ha='right')
plt.tight_layout()
plt.show()
```

# Parte 5: Histogramas — explorar distribuciones

Un **histograma** responde preguntas del tipo: *"¿cómo se distribuye esta variable? ¿Está concentrada, dispersa, tiene valores extremos?"*

A diferencia del barplot (que compara categorías), el histograma trabaja sobre **una sola variable numérica**.

### `sns.histplot()` — distribución del ARR

`sns.histplot()` agrupa los valores en *bins* (intervalos) y muestra cuántas observaciones caen en cada uno.

- `bins=` — cantidad de intervalos. Pocos bins: visión general. Muchos bins: más detalle, más ruido.
- `kde=True` — agrega una curva de densidad suavizada encima del histograma

**Pregunta de interpretación:** ¿la distribución del ARR es simétrica o está sesgada? ¿Qué implica eso si quisiéramos usar el *promedio* como métrica resumen?

```python
plt.figure(figsize=(10, 5))
sns.histplot(
    data=df,
    x='arr_millones',
    bins=25,
    kde=True,
    color='steelblue'
)
plt.title('Distribución del ARR (Ingresos Recurrentes Anuales)', fontsize=14)
plt.xlabel('ARR (millones USD)')
plt.ylabel('Cantidad de startups')
plt.tight_layout()
plt.show()
```

### `sns.histplot()` — distribución del churn mensual

El churn (% de clientes que se van por mes) es una métrica crítica para cualquier startup. Exploramos su distribución.

**Nota:** como `churn_mensual_pct` tiene nulos, seaborn los ignora automáticamente al graficar.

**Pregunta de interpretación:** mirando la distribución, ¿hay startups con un churn preocupante? ¿A partir de qué valor considerarían que el churn es "alto" para una startup LATAM?

```python
plt.figure(figsize=(10, 5))
sns.histplot(
    data=df,
    x='churn_mensual_pct',
    bins=20,
    kde=True,
    color='tomato'
)
plt.title('Distribución del Churn Mensual (%)', fontsize=14)
plt.xlabel('Churn mensual (%)')
plt.ylabel('Cantidad de startups')
plt.axvline(df['churn_mensual_pct'].median(), color='black', linestyle='--', label=f"Mediana: {df['churn_mensual_pct'].median():.1f}%")
plt.legend()
plt.tight_layout()
plt.show()
```

### `sns.histplot()` — distribución del NPS por modelo de negocio

Podemos usar `hue=` para superponer distribuciones de distintos grupos en el mismo histograma.

**Pregunta:** ¿esperaban que B2B y B2C tuvieran distribuciones de NPS diferentes? ¿Por qué o por qué no?

```python
df_nps = df[df['modelo_negocio'].isin(['B2B', 'B2C'])]  # excluimos B2B2C para simplificar

plt.figure(figsize=(10, 5))
sns.histplot(
    data=df_nps,
    x='nps_score',
    hue='modelo_negocio',
    bins=20,
    kde=True,
    alpha=0.5,
    palette=['steelblue', 'tomato']
)
plt.title('Distribución del NPS Score — B2B vs B2C', fontsize=14)
plt.xlabel('NPS Score')
plt.ylabel('Cantidad de startups')
plt.tight_layout()
plt.show()
```
